```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Salary Data.csv')
df.head()
```

```python
plt.scatter(df['Experience Years'], df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.show()
```

```python
x  = df.drop('Salary', axis=1)
y = df['Salary']
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
```

```python
df['predicted salary'] = reg.predict(x)
df.head(20)
```

```python
plt.plot(x,df['predicted salary'], color='red')
plt.scatter(df['Experience Years'], df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.show()
```

```python
reg.predict([[4.2]])
```

```python
score = reg.score(x,y)
print(score)
```

```python
df ['loss'] = df['Salary'] - df['predicted salary']
```

```python
df.head()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mse = mean_squared_error(df['Salary'], df['predicted salary'])
mae = mean_absolute_error(df['Salary'], df['predicted salary'])
```
